<h1>
    Documentation, the ML phase of the project so far:
</h1>

<h4>
    <ul>
        <li>
            <h4> 
            We started with the idea of simply <em>simulating</em> data from a cloud server, instead of using a direct web API, since the inclusion and process of exposing the API itself is very complex, so to run the model on simulated data is a lot 
            easier. 
            </h4>
        </li> 
        <li>
            <h4> 
           The Z score is way to determine outliers/anomalies in a given dataset, plotting the z score derived from the simulated 
            dataset clearly shows spikes in data, this can be utilised to train a machine learning model to actually predict future anomalies. 
            </h4>
        </li> 
        <li>
            <h4> 
           At first, we decided to test on a linear regression model. Although the model worked, it wasn't suitable for the task 
                of classifying clustered data as seen from a working server data stream.
            </h4>
        </li> 
        <li>
            <h4> 
           In order to fix this and to further improve prediction accuracy, we decided to switch to a logistic regression model, 
            this further boosted prediction capabilities, since a logistic regression model is best designed to handle clustered scatterpoints. 
            </h4>
        </li> 


</h4>
